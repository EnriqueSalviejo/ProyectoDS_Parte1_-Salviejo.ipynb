### WeatherAUS

### Este conjunto de datos comprende aproximadamente 10 años de observaciones meteorológicas diarias desde numerosos lugares de Australia.

Descripción:

* Date: Fecha de observación.
* Location: El nombre común de la ubicación de la estación meteorológica.
* MinTemp: Temperatura mínima en grados centígrados.
* MaxTemp: Temperatura máxima en grados centígrados.
* Rainfall: La cantidad de lluvia registrada durante el día en mm.
* Evaporation: La denominada evaporación del recipiente Clase A (mm) en el período de 24 horas a las 9 a.m.
* Sunshine: El número de horas de sol brillante en el día.
* WindGustDir: La dirección de la ráfaga de viento más fuerte en las 24 horas hasta la medianoche.
* WindGustSpeed: La velocidad (km/h) de la ráfaga de viento más fuerte en las 24 horas previas a la medianoche.
* WindDir9am: Dirección del viento a las 9 a.m.
* WindDir3pm: Dirección del viento a las 3 p.m.
* WindSpeed9am: Velocidad del viento (km/h) promediada durante 10 minutos antes de las 9 a.m.
* WindSpeed3pm: Velocidad del viento (km/h) promediada durante 10 minutos antes de las 3 p.m.
* Humidity9am: Humedad (porcentaje) a las 9 a.m.
* Humidity3pm: Humedad (porcentaje) a las 3 p.m.
* Pressure9am: La presión atmosférica (hpa) se redujo al nivel medio del mar a las 9 a.m.
* Pressure3pm: La presión atmosférica (hpa) se redujo al nivel medio del mar a las 3 p.m.
* Cloud9am: Fracción del cielo oscurecida por las nubes a las 9:00 horas. Esto se mide en "oktas", que son una unidad de octavos.
* Cloud3pm: Fracción del cielo oscurecida por las nubes a las 15:00 horas. Esto se mide en "oktas", que son una unidad de octavos.
* Temp9am: Temperatura (grados C) a las 9 a.m.
* Temp3pm: Temperatura (grados C) a las 3 p.m.
* RainToday: Booleano, 1 si la precipitación (mm) en las 24 horas hasta las 9 a. m. excede 1 mm; en caso contrario, 0.
* RainTomorrow: La cantidad de lluvia del día siguiente en mm. Se utiliza para crear la variable de respuesta RainTomorrow. Una especie de medida del "riesgo". Si la lluvia de ese día fue de 1 mm o más.

```python
# Identificar las columnas numéricas
numeric_cols = df_weather_aus.select_dtypes(include=['number']).columns

# Rellenar los valores nulos solo en las columnas numéricas
df_weather_aus[numeric_cols] = df_weather_aus[numeric_cols].fillna(df_weather_aus[numeric_cols].median())
df_weather_aus.head()
```

```python
df_weather_aus['Date'] = pd.to_datetime(df_weather_aus['Date'])
```
